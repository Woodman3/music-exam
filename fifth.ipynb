```python
from fractions import Fraction

def fifth_cycle(max_n=60):
    k =0
    r = Fraction(1,1)

    for n in range(1, max_n+1):

        r *= Fraction(3,2)
        
        # 折回一个八度
        while r >= 2:
            r /= 2
            k += 1
        while r < 1:
            r *= 2
            k -= 1

        error = float(r - 1)
        
        if error < 0.02:
            print(
                f"n={n:3d}  "
                f"ratio={r}  "
                f"value={float(r):.12f}  "
                f"error={error:.12f}  "
                f"k={k:3d}"
            )


fifth_cycle(500)
```
